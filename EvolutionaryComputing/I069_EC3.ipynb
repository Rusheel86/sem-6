#Using Backtracking

```python
def printSolution(board, N):
    for i in range(N):
        for j in range(N):
            print (board[i][j],end=' ')
        print()
```

```python
def isSafe(board, row, col, N):

    for i in range(col):
        if board[row][i] == 1:
            return False

    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    for i, j in zip(range(row, N, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    return True
```

```python
def solveNQUtil(board, col, N):

    if col >= N:
        return True

    for i in range(N):
        if isSafe(board, i, col,N):
            board[i][col] = 1
            if solveNQUtil(board, col + 1,N) == True:
                return True
            board[i][col] = 0

    return False
```

```python
def solveNQ(N):
    board = [[0 for _ in range(N)] for _ in range(N)]


    if solveNQUtil(board, 0, N) == False:
        print ("Solution does not exist")
        return False

    printSolution(board,N)
    return True
```

```python
N = int(input("Enter the number of queens: "))
solveNQ(N)
```

#Using EC

```python
import random
```

```python
def get_inputs():
  print("Enter number of queens")
  n = input("n")
  print("Enter number of iterations")
  iterations = input("iterations")
  print("Enter population size")
  population_size = input("population_size")
  return n, iterations, population_size
```

```python
def create_individual(n):
    return [random.randint(0, n - 1) for _ in range(n)]
```

```python

def calculate_fitness(individual):
    n = len(individual)
    max_fitness = (n * (n - 1)) // 2
    collisions = 0
    for i in range(n):
        for j in range(i + 1, n):
            if individual[i] == individual[j]:
                collisions += 1
            elif abs(individual[i] - individual[j]) == abs(i - j):
                collisions += 1
    return max_fitness - collisions
```

```python
def crossover(parent1, parent2):
    n = len(parent1)
    crossover_point = random.randint(1, n - 1)
    return parent1[:crossover_point] + parent2[crossover_point:]
```

```python
MUTATION_RATE = 0.5
```

```python
def mutate(individual):
    if random.random() < MUTATION_RATE:
        n = len(individual)
        idx = random.randint(0, n - 1)
        val = random.randint(0, n - 1)
        individual[idx] = val
    return individual
```

```python
def solve_genetic_n_queens(n, max_generations, pop_size):

    population = [create_individual(n) for _ in range(pop_size)]
    generation = 0
    max_possible_fitness = (n * (n - 1)) // 2

    while generation < max_generations:

        population.sort(key=calculate_fitness, reverse=True)

        if calculate_fitness(population[0]) == max_possible_fitness:
            print(f"Solution found in generation {generation}!")
            return population[0]

        parents = population[:pop_size // 2]

        next_generation = []
        next_generation.append(parents[0])
        while len(next_generation) < pop_size:
            parent1 = random.choice(parents)
            parent2 = random.choice(parents)
            child = mutate(crossover(parent1, parent2))
            next_generation.append(child)
        population = next_generation
        generation += 1

    print("Max generations reached.")
    return population[0]
```

```python
def print_board(solution):
    n = len(solution)
    for row in range(n):
        line = ""
        for col in range(n):
            if solution[col] == row:
                line += "Q "
            else:
                line += ". "
        print(line)
```

```python
if __name__ == "__main__":
    try:

        n_input, iter_input, pop_input = get_inputs()

        N = int(n_input)
        GENERATIONS = int(iter_input)
        POP_SIZE = int(pop_input)

        solution = solve_genetic_n_queens(N, GENERATIONS, POP_SIZE)
        print("\nBest Individual:", solution)
        print("Fitness:", calculate_fitness(solution))
        print("\nBoard Visualization:")
        print_board(solution)

    except ValueError:
        print("\nError: Please enter valid integers for all inputs.")
```
