4.	Given is the map of Australia. Perform the following task-

    a.	To color each region either red, green, or blue in such a way that no two neighboring regions have the same color

    b.	To formulate this as a CSP

    c.	Identify variables, domain and constraints

    d.	Implement this CSP in Python

    image.png

```python
states = ["WA","NT","SA","QL","NSW","V","TA"]
```

```python
colors = ["Red", "Green", "Blue"]
```

```python
neighbors = {
    "WA": ["NT", "SA"],
    "NT": ["WA", "SA", "QLD"],
    "SA": ["WA", "NT", "QLD", "NSW", "VIC"],
    "QL": ["NT", "SA", "NSW"],
    "NSW": ["QLD", "SA", "VIC"],
    "V": ["SA", "NSW"],
    "TA": []
}
```

```python
assignment = {}
```

```python
def solve(index):
    if index == len(states):
        return True

    state = states[index]

    for color in colors:
        if all(assignment.get(n) != color for n in neighbors[state]):
            assignment[state] = color
            if solve(index + 1):
                return True
            assignment[state] = None
    return False
```

```python
if solve(0):
    print("Solution found:", assignment)
else:
    print("No solution exists.")
```
